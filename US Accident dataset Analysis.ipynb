#### Start Time

```python
#change to date time
df.Start_Time=pd.to_datetime(df.Start_Time)
```

### Is Start_Time has any impact on the accident ratios?

```python
sns.distplot(df.Start_Time.dt.hour,bins=24,kde=False,norm_hist=True)
```

### Which weekdays have more number of accident?

```python
sns.distplot(df.Start_Time.dt.day_of_week,bins=7,kde=False,norm_hist=True)
```

### Which time of day Sunday is having more accicents??

```python
#sunday accident distribution
sundays_start_time=df.Start_Time[df.Start_Time.dt.dayofweek==6]
sns.distplot(sundays_start_time.dt.hour,bins=24,kde=False,norm_hist=True)
```

### Accident distribution for Monday

```python
# visualise the accident distribution of Monday for comparing with Sunday
monday_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 0]
sns.distplot(monday_start_time.dt.hour, bins=24, kde=False, norm_hist=True)
```

### Which months of year see more number of accidents?

```python
sns.distplot(sundays_start_time.dt.month,bins=12, kde=False, norm_hist=True)
```

### Is there any Covid-19 lockdown has any impression on the accidents?

```python
# 2020 year data visualisation
df_2020=df[df.Start_Time.dt.year==2020]
sns.distplot(df_2020.Start_Time.dt.month, bins=12, kde=False, norm_hist=True)
```

```python
#compare with 2019 to get better understanding
df_2019=df[df.Start_Time.dt.year==2019]
sns.distplot(df_2019.Start_Time.dt.month, bins=12, kde=False, norm_hist=True)
```
