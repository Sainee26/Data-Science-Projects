# Which set of moves will get the best ratio of attack power to SP spent?¶

```python
movelist.head()
```

```python
movelist.rename(columns={'SP Cost':'sp_cost'}, inplace=True)
```

```python
#to recheck
movelist.head()
```

```python
movelist["best_ratio"] = round(movelist.Power/movelist.sp_cost)
movelist.sort_values("best_ratio", ascending=False)
```

Answer: We can see that Digimon that have the move 'Heavy Strike I' is good because it has a really efficient damage with low SP cost and somewhat good Power.

# Are different types and attributes evenly represented across stages?¶

```python
plt.figure(figsize=(12,4))
sns.countplot(hue=digimon['Type'], x=digimon['Stage'])
plt.show()
```

ANSWER: *Here as we can see that Digimon Types are not really represented across stages. Digimon's type started as Free-type in Baby and In-Training stages, then as they reached their last stage, they branches evenly towards the other types, with only some Digimon retaining their Free-type. As they keep growing to other higher stages (besides Ultra and Armor stage, as they are special stage), the Digimon's type tends to be more of a Virus-type.*
